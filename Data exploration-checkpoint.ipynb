## Data exploration on Fantasy Football stats

```python
import requests;
import pandas as pd;
```

```python
#Following https://towardsdatascience.com/fantasy-premier-league-value-analysis-python-tutorial-using-the-fpl-api-8031edfe9910 to obtain fpl data via API

url = 'https://fantasy.premierleague.com/api/bootstrap-static/'

r = requests.get(url)

json = r.json()

print(json.keys())

elements_df = pd.DataFrame(json['elements'])
elements_types_df = pd.DataFrame(json['element_types'])
element_stats_df = pd.DataFrame(json['element_stats'])
teams_df = pd.DataFrame(json['teams'])
events_df = pd.DataFrame(json['events'])
```

```python
#Chaning settings to remove limit from number of columns displayed
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
elements_df.head(5)
```

```python
#Replacing NaN values with 0
elements_df.fillna(0,inplace=True)
```

```python
teams_df.head(5)
```

```python
#Renaming 'id' column in teams_df to 'team'
teams_df=teams_df.rename(columns={'id':'team'})
```

```python
#Merging elements_df with teams_df to add team name to data
elements_df=pd.merge(elements_df,teams_df[['team','name']],on='team')
```

```python
elements_df.head(5)
```

```python
#Creating filter to filter out individuals with 0 change of playing next round
player_filter = (elements_df['chance_of_playing_next_round']>0)
```

```python
#Adjusting data frame with filter in place
elements_df=elements_df[player_filter]
```

```python
#Dataframe now only contains players likely to play
elements_df.head(5)
```

```python
elements_df.dtypes
```

```python
#Converting columns ep_next,ep_this,points_per_game to float64 type
elements_df['ep_next']=elements_df['ep_next'].astype(float)
elements_df['ep_this']=elements_df['ep_this'].astype(float)
elements_df['points_per_game']=elements_df['points_per_game'].astype(float)
```

```python
Focus_Teams = ['Arsenal','Man City','Newcastle','Spurs']
```

```python
#Creating new dataframe that only focuses on specific teams
Team_Filter = elements_df[elements_df['name'].isin(Focus_Teams)]
```

```python
Team_Filter.head(50)
```

```python
#Creating histogram to view expected points distribution for each of top 4's next match
Team_Filter['ep_next'].hist(by=Team_Filter['name'])
```
