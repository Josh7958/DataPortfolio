## Importing libraries and preparing data set

```python
import numpy as np 
import pandas as pd
import matplotlib.pylab as plt 
import matplotlib.pyplot as plot
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

POL = pd.read_csv(r'C:\Users\[WideWorldImporters].[Purchasing].[PurchaseOrderLines].csv')
PO = pd.read_csv(r'C:\Users\[WideWorldImporters].[Purchasing].[PurchaseOrders].csv')
```

```python
#Preparing new dataframes based on the columns we need
PO_dates = PO[['PurchaseOrderID','OrderDate']]
Order = POL[['PurchaseOrderID','StockItemID','OrderedOuters']]

#Merging two dataframes together
Order_data = pd.merge(Order,PO_dates,on='PurchaseOrderID')
```

```python
#Transforming Date column to beginning of month
Order_data['Month'] = pd.to_datetime(Order_data["OrderDate"]).dt.floor('d') + pd.offsets.MonthEnd(0) - pd.offsets.MonthBegin(1)
```

```python
print(Order_data)
```

```python
#Transforming month column to date format
Order_data['Month']=Order_data['Month'].dt.date
```

```python
#Selecting the two columns necessary for plotting/forecasting
Order_data=Order_data[["OrderedOuters","Month"]]
```

```python
#Grouping data by month value
Order_data=Order_data.groupby(Order_data.Month).sum()
```

```python
print(Order_data.head(5))
```

```python
plt.plot(Order_data)
```

```python
#Setting date limits for filtering of data - to focus on smaller subset
startdate = pd.to_datetime("2013-01-01").date()
enddate = pd.to_datetime("2015-12-31").date()
```

```python
#Filtering data on which forecasting functions will be trained based on start and end dates previously defined
Train_data = Order_data[(Order_data.index >= startdate) & (Order_data.index <= enddate)]
```

```python
Train_data=Train_data[['OrderedOuters']]
```

```python
Train_data = Train_data.groupby(Train_data.index).sum()
```

```python
x = Order_data.index
plt.xticks(rotation=90)
gr=plt.plot(Train_data)
```

## Statistical tests carried out to check suitability of data for forecasting

```python
#Calculating rolling mean and standard deviation of data for later suitability checks for forecasting
Rollingmean=Train_data.rolling(window=12).mean()
Rollingstd=Train_data.rolling(window=12).std()
```

```python
gr=plt.plot(Train_data, label='Orders')
RM_Plot = plt.plot(Rollingmean,label='Mean')
RS_Plot = plt.plot(Rollingstd,label='Std')
plt.legend()
plt.xlabel('Date')
plt.ylabel('Orders')
plt.xticks(rotation=90)
#for label in gr.get_xaxis().get_ticklabels()[::2]:
 #   label.set_visible(False)
plt.show()
```

```python
#p-value is too large to reject null hypothesis. Additionally, critical value is not greater than the test statistic

from statsmodels.tsa.stattools import adfuller

dftest=adfuller(Train_data,autolag = 'AIC')
dfoutput = pd.Series(dftest[0:4],index=['Test Statistic','p-value','#lags used','Number of observations used'])
for key, value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key]=value
    
    print(dfoutput)
```

## Attempting various data transformations to make data stationary and thus suitable for forecasting function

```python
Train_data_logscale = np.log(Train_data)
Train_data_logscale = Train_data_logscale.astype(float)
print(Train_data_logscale["OrderedOuters"].dtype)
gr_log = plt.plot(Train_data_logscale)
plt.xlabel('Date')
plt.ylabel('Orders (log scale)') 
plt.xticks(rotation=90)
for label in gr_log.get_xaxis().get_ticklabels()[::2]:
    label.set_visible(False)
   
plt.show()    
```

```python
#Plot logarithmic values to demonstrate non-stationarity of data
MovingAverage = Train_data_logscale.rolling(window=12).mean()
MovingSTD = Train_data_logscale.rolling(window=12).std()
plt.plot(MovingAverage,label='moving average')
plt.plot(MovingSTD, label='moving std')
plt.xticks(rotation=90)
plt.xlabel('Date')
plt.legend()
plt.show()
```

```python
#Find difference between logarithmic data and moving average
Datalogscaleminusmovingaverage = Train_data_logscale - MovingAverage

Datalogscaleminusmovingaverage.dropna(inplace=True)
Datalogscaleminusmovingaverage.head(10)
```

```python
def test_4_stationarity(timeseries):
    #plot rolling statistics
    movinglogavg=timeseries.rolling(window=6).mean()
    movinglogstd=timeseries.rolling(window=6).std()
    
    #plotting original data
    orig=plt.plot(timeseries, color = 'blue', label = 'Original')
    #plot mean
    mean=plt.plot(movinglogavg, color='red', label='mean')
    #plot std
    std=plt.plot(movinglogstd, color ='green', label='std')
    plt.xticks(rotation=90)
    
    plt.legend(loc='best')
    plt.show(block=False)
    
  #perform dickey-fuller test
    print('results of Dickey-Fuller test')
    dftest=adfuller(timeseries,autolag='AIC')
    dfoutput=pd.Series(dftest[0:4], index=['Test Statistic','p-value','#lags used','Number of observations used'])
    for key,value in dftest[4].items():
        dfoutput['Critical value (%s)'%key]=value
    print(dfoutput)
```

```python
test_4_stationarity(Datalogscaleminusmovingaverage)
```

```python
exponentialdecayweightedaverage = Train_data_logscale.ewm(halflife=12, min_periods=0, adjust=True).mean()
plt.plot(Train_data_logscale)
plt.plot(exponentialdecayweightedaverage,color='red')
plt.show()
    
test_4_stationarity(exponentialdecayweightedaverage)
```

```python
Datalogscaleminusmovingaverage = Train_data_logscale - exponentialdecayweightedaverage
test_4_stationarity(Datalogscaleminusmovingaverage)
```

```python
Datasetlogdiffshifting = Train_data_logscale - Train_data_logscale.shift()
plt.xticks(rotation=90)
plt.plot(Datasetlogdiffshifting)

Datasetlogdiffshifting.dropna(inplace=True)
test_4_stationarity(Datasetlogdiffshifting)
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = seasonal_decompose(Train_data_logscale,model='additive', period=6)

trend = decomposition.trend
seasonal = decomposition.seasonal
residual = decomposition.resid


residual = residual.astype(float)


decomposedlogdata=residual
decomposedlogdata.dropna(inplace=True)
test_4_stationarity(decomposedlogdata)
```

```python
#ACF and PACF plots

from statsmodels.tsa.stattools import acf,pacf

lag_acf = acf(exponentialdecayweightedaverage, nlags=15)
lag_pacf = pacf(exponentialdecayweightedaverage,nlags=15, method='ols')

#plot acf
plt.subplot(121)
plt.plot(lag_acf)
plt.axhline(y=0, linestyle = '--', color = 'gray')
plt.axhline(y=-1.96/np.sqrt(len(Datasetlogdiffshifting)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(Datasetlogdiffshifting)), linestyle='--', color='gray')

plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0, linestyle = '--', color = 'gray')
plt.axhline(y=-1.96/np.sqrt(len(Datasetlogdiffshifting)), linestyle='--', color='gray')
plt.axhline(y=1.96/np.sqrt(len(Datasetlogdiffshifting)), linestyle='--', color='gray')
plt.show()
```

## Importing libraries to forecast 

```python
from statsmodels.tsa.arima_model import ARIMA

model=ARIMA(Datasetlogdiffshifting, order=(1,1,2))
results_AR = model.fit(disp=-1)
plt.xticks(rotation=90)
plt.plot(Datasetlogdiffshifting)
plt.plot(results_AR.fittedvalues, color='red')
plt.title('RSS:%.4f'% sum((results_AR.fittedvalues-Datasetlogdiffshifting.OrderedOuters)**2))
print('Plotting AR model')
plt.show()
(results_AR.fittedvalues.head()-Datasetlogdiffshifting.OrderedOuters.head())**2
```
